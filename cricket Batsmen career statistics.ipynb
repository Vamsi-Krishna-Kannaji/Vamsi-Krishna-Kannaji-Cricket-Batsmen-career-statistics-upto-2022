```python
df['Start_Year'] = df['Span'].str.split('-').str[0].astype(int)
df['End_Year'] = df['Span'].str.split('-').str[1].astype(int)

df['Start_Date'] = pd.to_datetime(df['Start_Year'], format='%Y')
df['End_Date'] = pd.to_datetime(df['End_Year'], format='%Y')

df['Duration_Years'] = (df['End_Date'] - df['Start_Date']).dt.days / 365.25

# Drop intermediate columns
df.drop(['Start_Year', 'End_Year', 'Start_Date', 'End_Date'], axis=1, inplace=True)

print(df)
```
